```python
# Contagem das áreas correlacionadas mais procuradas
area_count = df['Área correlacionada mais procurada'].value_counts()
```

```python
# Gráfico de barras das áreas mais procuradas
plt.figure(figsize=(10, 6))
sns.barplot(x=area_count.index, y=area_count.values, palette='viridis')
plt.title('Contagem das Áreas Correlacionadas Mais Procuradas')
plt.xlabel('Áreas')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```

```python
# Análise da população que estuda TI por capital
plt.figure(figsize=(12, 8))
sns.barplot(x='População que estuda TI', y='Capital', data=df, palette='viridis')
plt.title('População que estuda TI por Capital')
plt.xlabel('População que estuda TI')
plt.ylabel('Capital')
plt.show()
```

```python
# Analisando as áreas correlacionadas estudadas
areas = df['Áreas correlacionadas estudadas'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
areas_count = areas.value_counts()
```

```python
# Gráfico de barras das áreas correlacionadas estudadas
plt.figure(figsize=(10, 6))
sns.barplot(x=areas_count.values, y=areas_count.index, palette='viridis')
plt.title('Contagem das Áreas Correlacionadas Estudadas')
plt.xlabel('Contagem')
plt.ylabel('Áreas')
plt.show()
```
